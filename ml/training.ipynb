```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
links=pd.read_csv("/content/links.csv")
links.head(5)
```

```python
ratings=pd.read_csv("/content/ratings.csv")
ratings.head(5)
```

```python
ratings['userId'].nunique()
ratings['movieId'].nunique()
```

```python
tags=pd.read_csv("/content/tags.csv")
tags.head()
```

```python
movies=pd.read_csv("/content/movies.csv")
movies.head()
```

```python
df=pd.merge(ratings,movies,on='movieId')
```

```python
df.head(5)
```

```python
user_movie=df.pivot_table(
    index='userId',
    columns='title',
    values='rating'
)
```

```python
user_movie=user_movie.fillna(0)
print(user_movie)
```

```python
similarity=cosine_similarity(user_movie)

similarity_df=pd.DataFrame(similarity, index=user_movie.index,
                          columns=user_movie.index)
print(similarity_df)
```

```python
user_id=1
```

```python
similar_users=similarity_df[user_id].sort_values(ascending=False)[1:6]
print(similar_users)
```

```python
movie=user_movie.loc[similar_users.index]
print(movie.head())
```

```python
recommended = movie.mean().sort_values(ascending=False)

print(recommended.head())
```

```python
watched = user_movie.loc[user_id]

recommended = recommended[watched == 0]
```

```python
print(recommended.head(5))
```
